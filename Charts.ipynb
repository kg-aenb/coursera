```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np

xdata = np.random.rand(100)*8
ydata = np.random.rand(100)*12
```

# **PyPlot functions**

```python
# line plot
plt.plot(np.sort(xdata), ydata)
```

```python
# scatter plot
plt.scatter(xdata, ydata)
```

```python
# histogram
plt.hist(xdata, edgecolor='black') #, bins=5)
# a lot of options for bins and other parameters ;)
```

```python
# bar plot
plt.bar(list('X'+ str(i) for i in range(1,11)), ydata[:10], edgecolor='red')
```

```python
#  Pseudo Color Plot
mm = np.outer(np.sort(xdata[:8]), ydata[:12])
plt.pcolor(mm, cmap='RdBu')
plt.colorbar()
```

# **Seaborn functions**

```python
#  Box and whisker plot

sns.boxplot([xdata[:8], ydata[:30]])

# Also a dataframe (and a lot of other types) may be used as a source of data
```

```python
# A regression plot draws a scatter plot of two variables, x and y,
# and THEN fits the regression model and plots the resulting regression line
# along with a 95% confidence interval for that regression

sns.regplot(x = np.sort(xdata[:20]),y = np.sort(ydata[:20]))

# Accepts dataframe and other types of input data
# sns.regplot(x = 'header X',y = 'headerY',data= df)
```

```python
# A residual plot is used to display the quality of polynomial regression
# This function will REGRESS y on x as a polynomial regression and
# THEN draw a scatterplot of the residuals
sns.residplot(x=np.sort(xdata[:20]), y=np.sort(ydata[:20]))

# Again, may work with dataframes and other types of input data
```

```python
# Distribution Plot
sns.displot(xdata, color='r', kind="kde");  # kind = 'hist', 
# as always, a lot of options - see next example below
```

```python
# 2+ distributions on the same plot - single liner
width = 8
height = 1
# plt.figure(figsize=(width, height))

# mypal = sns.color_palette(['red','blue'])  # set own colors instead of defaults 
# mypal = ['red','blue']

sns.displot({'X data':xdata, 'Y data':ydata}, palette=['r','b'], kind="kde")

fig = plt.gcf()
fig.set_size_inches(width, height)
plt.title('Example of KDE chart')
plt.xlabel('Simple X axis')
plt.ylabel('Example of Y axis') # instead of defaut 'Density'
plt.legend()
```

```python
# two distributions on a single plot - another example

width = 7
height = 2
fig = plt.gcf()
fig.set_size_inches(width, height);
# mypal = sns.color_palette(['red','blue'])  # set own colors instead of defaults 

ax1 = sns.kdeplot(xdata, color="r", label='X data in red', legend=True)
ax2 = sns.kdeplot(ydata, color="b", label='Y data in blue', ax=ax1, legend=True)
# sns.displot({'X data':xdata, 'Y data':ydata}, palette=mypal, kind="kde")

plt.title('Example of KDE chart')
plt.xlabel('Simple X axis')
plt.ylabel('Example of Y axis') # instead of defaut 'Density'
plt.legend();
```
